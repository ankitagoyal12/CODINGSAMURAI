```python
import pandas as pd

df = pd.read_csv('sales_data.csv')
df.head()  # shows the first 5 rows of your dataset
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

# Sum sales by product
sales_by_product = df.groupby('Product')['Sales'].sum()

# Bar chart
sales_by_product.plot(kind='bar', color='Purple')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
# Sum quantity by category
quantity_by_category = df.groupby('Category')['Quantity'].sum()

# Pie chart
quantity_by_category.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Quantity Sold by Category')
plt.ylabel('')  # removes y-label for better look
plt.show()
```
